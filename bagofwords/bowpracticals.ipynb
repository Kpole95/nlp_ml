```python
## text cleaning

import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
porter_stemmer=PorterStemmer()
```

```python
## PorterS Stemming

corpus_porter=[]

for i in range(0, len(messages)):
    review=re.sub("[^a-zA-z]",' ', messages['message'][i]) # remove special characters
    review=review.lower() # lower teh words
    review=review.split()  # slipt and get a list of words
    review=[porter_stemmer.stem(word) for word in review if not word in stopwords.words('english')] #
    review=" ".join(review)
    corpus_porter.append(review)
```

```python
corpus_porter
```

```python
## snowball stemming

from nltk.stem import SnowballStemmer
snowball_stemmer=SnowballStemmer("english")
```

```python
corpus_snowball=[]

for i in range(0, len(messages)):
    review=re.sub("[^a-zA-z]",' ', messages['message'][i]) # remove special characters
    review=review.lower() # lower teh words
    review=review.split()  # slipt and get a list of words
    review=[snowball_stemmer.stem(word) for word in review if not word in stopwords.words('english')] #
    review=" ".join(review)
    corpus_snowball.append(review)
```

```python
## lemmatization

from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus_lematizer=[]

for i in range(0, len(messages)):
    review=re.sub("[^a-zA-z]",' ', messages['message'][i]) # remove special characters
    review=review.lower() # lower teh words
    review=review.split()  # slipt and get a list of words
    review=[lemmatizer.lemmatize(word, pos='v') for word in review if not word in stopwords.words('english')] #
    review=" ".join(review)
    corpus_lematizer.append(review)
```

```python
corpus_lematizer
```
