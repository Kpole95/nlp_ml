## Peprocessing And Cleaning

```python
# data cleaning 
import re
import nltk
nltk.download('stopwords')
```

```python
# porter stemmer

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
porter_stemmer=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[porter_stemmer.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Train Test Split

```python
## output features

y=pd.get_dummies(messages['target'])
y=y.iloc[:,0].values
```

```python
## train, test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(corpus, y, test_size=0.20, random_state=42)
```

## TF-IDF (Sentences--->vectors) {Preventing Data Leakage}

```python
## BOW Model (TfidfVectorizer)

from sklearn.feature_extraction.text import TfidfVectorizer
tf_vec_porter=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train), len(y_train)
```

```python
# independent features

X_train=tf_vec_porter.fit_transform(X_train).toarray()
X_test=tf_vec_porter.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
    formatter= dict(float=lambda x: "%.3g" % x))

X_train
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
tf_vec_porter.vocabulary_
```

## Train the TF-IDF model with Porter Stemmer

```python
from sklearn.naive_bayes import MultinomialNB
tfidf_model_porter = MultinomialNB().fit(X_train, y_train)
```

```python
# prediction
y_pred=tfidf_model_porter.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## Creating the TF-IDF Model with Lemmatation

```python
# data cleaning 
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]

for i in range(0, len(messages)):
    review=re.sub("[^a-zA-z]",' ', messages['message'][i]) # remove special characters
    review=review.lower() # lower teh words
    review=review.split()  # slipt and get a list of words
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')] #
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

## Train, test split

```python
## output features

y=pd.get_dummies(messages['target'])
y=y.iloc[:,0].values
```

```python
## train, test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(corpus, y, test_size=0.20, random_state=42)
```

## TF-IDF (Sentences--->vectors) {Preventing Data Leakage}

```python
## BOW Model (TfidfVectorizer)

from sklearn.feature_extraction.text import TfidfVectorizer
tf_vec_lemma=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train), len(y_train)
```

```python
## independent features

X_train=tf_vec_lemma.fit_transform(X_train).toarray()
X_test=tf_vec_lemma.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
    formatter= dict(float=lambda x: "%.3g" % x))

X_train
```

## Train the TF-IDF Model with Lemmatation

```python
from sklearn.naive_bayes import MultinomialNB
tf_model_lemma=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=tf_model_lemma.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
