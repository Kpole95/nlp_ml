```python
import pandas as pd
import re
# messages = pd.read_csv(r"E:\learning\krishudemy\nlpml\bagofwords\spam.csv", encoding='latin1')
messages=pd.read_csv(r"E:\learning\krishudemy\nlpml\spamhap\cleaned_spam_data.csv", encoding='latin1')
```

```python
messages.shape
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review] # no stopwords 
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess # converts a doc into a list of lowercase tokens and ignores token which are too short and long.
```

```python
words=[]
for sentences in corpus:
    sent_token=sent_tokenize(sentences)
    for sentences in sent_token:
        words.append(simple_preprocess(sentences))
```

```python
## Train the Word2Vec Model from Scratch

import gensim
model=gensim.models.Word2Vec(words)
```

```python
import numpy as np
from tqdm import tqdm

def avg_word2vec(doc, model):
    """
    Calculates the average Word2Vec vector for a document (list of words).
    Returns a zero vector if no words are in the model's vocabulary.
    """
    # Get all word vectors for words that exist in the model's vocabulary
    sentence_vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    
    # If no words were found in the vocabulary, return a zero vector of the correct size
    if not sentence_vectors:
        return np.zeros(model.wv.vector_size)
    
    # Otherwise, return the mean of the word vectors
    return np.mean(sentence_vectors, axis=0)

# Apply the corrected function to every sentence in your 'words' list.
# This creates a clean list where every element is a vector of the same size.
X = [avg_word2vec(sentence, model) for sentence in tqdm(words)]

# Now, you can safely convert this to a NumPy array
X_new = np.array(X)

# Check the final shape to confirm it's a proper 2D array (samples, features)
print("\nShape of the final feature matrix:", X_new.shape)
```
